```python
pilote_dans_écurie = resusts[["driverId","constructorId"]].drop_duplicates().sort_values("constructorId")
pilote_dans_écurie
```

```python
drive = pd.merge(drivers, pilote_dans_écurie, on='driverId')
constr = pd.merge(drive, constructeur, on='constructorId')
constr =constr[['driverId', 'surname', 'nationality_x', 'constructorId', 'name', 'nationality_y']]
constr["match"] = (constr["nationality_x"] == constr["nationality_y"]).astype(int)
nb_pilote = constr.groupby('constructorId').count()[['driverId']].reset_index()
nb_pilote2 = constr.groupby('constructorId').sum()[['match']]
nb_pilote
```

```python
average_constructor = construct_result.merge(races, on="raceId")
average_constructor = average_constructor[["year","constructorId","points"]].groupby(["year","constructorId"]).sum()
average_constructor = average_constructor.reset_index()
average_constructor = average_constructor[["constructorId","points"]].groupby("constructorId").mean()
average_constructor = pd.merge(average_constructor, constructeur, on='constructorId')
average_constructor = average_constructor[['name', 'points']].sort_values(by='points', ascending=False).reset_index(drop=True)
```

```python
nb_pilote3 = nb_pilote.join(nb_pilote2)
nb_pilote3['pourcentage']= (nb_pilote3['match']/nb_pilote3['driverId'])*100
nb_pilote3 = nb_pilote3.reset_index()
lol = pd.merge(nb_pilote3, constructeur, on='constructorId')
lol = lol.merge(average_constructor.reset_index(), on='constructorId')
lol['pourcentage-étranger'] = 100 - lol['pourcentage']
lol['non-match'] = lol['driverId']- lol['match']
lol =lol[['name', 'driverId', 'match', 'non-match', 'pourcentage', 'pourcentage-étranger', 'points']]
lol

```
